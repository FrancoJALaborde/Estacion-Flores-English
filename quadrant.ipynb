## Complete Quadrant

```python
import folium

upper_left_latitude = -31.432580
upper_left_longitude = -64.252421
lower_right_latitude = -31.451888
lower_right_longitude = -64.242889

center_latitude = (upper_left_latitude + lower_right_latitude) / 2
center_longitude = (upper_left_longitude + lower_right_longitude) / 2

m = folium.Map(location=[center_latitude, center_longitude], zoom_start=12)

rectangle = folium.Rectangle(
    bounds=[
        (upper_left_latitude, upper_left_longitude),
        (lower_right_latitude, lower_right_longitude),
    ],
    color="blue",
    fill=True,
    fill_color="#3186cc",
).add_to(m)

folium.Marker([center_latitude, center_longitude], popup="Center of Rectangle").add_to(
    m
)

m.save("quadrant_maps/full_quadrant_map.html")
m
```
